# Normalization

```python
alc = alcantarilla.drop(columns=['Date','Month','Hour'])
normalized_alc=(alc-alc.min())/(alc.max()-alc.min())
normalized_alc
```

```python
normalized_alc['SO2 (µg/m³N)'] = normalized_alc['SO2 (µg/m³N)'].where(normalized_alc['SO2 (µg/m³N)']<0.6)
normalized_alc['PM10 (µg/m³N)'] = normalized_alc['PM10 (µg/m³N)'].where(normalized_alc['PM10 (µg/m³N)']<0.5)
```

# Imputing missing values

```python
normalized_alc['Date'] = d[:-1]
normalized_alc['Date'] = pd.to_datetime(normalized_alc['Date'], format='%d-%m-%Y')
normalized_alc['Month'] = d[:-1].strftime("%m")
normalized_alc['Hour'] = d[:-1].strftime("%H")
normalized_alc = normalized_alc.set_index('Date')
normalized_alc
```

## kNN imputation

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5) 
After_imputation = imputer.fit_transform(normalized_alc)
alc_knn = pd.DataFrame(After_imputation, columns = ['NO (µg/m³N)','NO2 (µg/m³N)','SO2 (µg/m³N)','O3 (µg/m³N)','TMP (ºC)','HR (% H.R.)','NOX (µg/m³N)','DD (grados)','RS (W/m³)','VV (m/s)','PM10 (µg/m³N)','Month','Hour'])
alc_knn
```

```python
denormalized_alc=(alc_knn*(alc.max()-alc.min()))+alc.min()
denormalized_alc['Date'] = d[:-1]
denormalized_alc['Date'] = pd.to_datetime(denormalized_alc['Date'], format='%d-%m-%Y')
in_array = np.linspace(0, np.pi, 17520)
out_array = np.sin(in_array)
denormalized_alc['Month'] = 12*out_array
denormalized_alc['Hour'] = d[:-1].strftime("%H")
denormalized_alc.to_csv('alc.csv',index=False)
denormalized_alc
```
